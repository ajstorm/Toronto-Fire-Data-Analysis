```python
# Install and include prerequisites
!pip install --upgrade folium
!pip install --upgrade geopandas
import folium
import pandas as pd
import geopandas as gpd
from sklearn.cluster import KMeans
```

```python
# Get the fire data and load into a data frame
!wget -O fire_data.csv https://ckan0.cf.opendata.inter.prod-toronto.ca/download_resource/fa5c7de5-10f8-41cf-883a-9b30a67c7b56?format=csv
df = pd.read_csv("fire_data.csv")
```

```python
# Clean the data - Remove NaN values
map_df = df[df['Longitude'] < 0]
map_df = map_df[['Latitude', 'Longitude']]
```

```python
# Calculate the k-means clustering for 10 centroids
kmeans = KMeans(n_clusters=10).fit(map_df)
fire_centroids = kmeans.cluster_centers_
```

```python
# Map the centroids
folium_map = folium.Map(location=[43.7, -79.35],
                        zoom_start=12,
                        tiles="CartoDB positron")

for i in fire_centroids:
    folium.Marker(location=i, icon=folium.Icon(color='red', icon='fire', prefix='fa')).add_to(folium_map)
folium_map
```

```python
# Calculate the k-means clustering for 30 centroids
kmeans = KMeans(n_clusters=30).fit(map_df)
fire_centroids = kmeans.cluster_centers_
```

```python
# Map the centroids
folium_map = folium.Map(location=[43.7, -79.35],
                        zoom_start=12,
                        tiles="CartoDB positron")

for i in fire_centroids:
    folium.Marker(location=i, icon=folium.Icon(color='red', icon='fire', prefix='fa')).add_to(folium_map)
folium_map
```
